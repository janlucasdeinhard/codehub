# Covid Forecast
This project attempts to forecast the total number of expected Covid cases per country, based on the exponential growth observed so far. The forecast is based on an exponential rise - saturation model of disease spread.

```python
import numpy as np
from numpy import random as rnd
from matplotlib import pyplot as plt
import warnings,os,datetime,time,sys

import pandas as pd
```

```python
df = pd.read_csv('../../../TF_Data/Dropbox/TF_data/COVID19/time_series_covid_19_confirmed.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Province/State','Lat','Long']).transpose()
```

```python
df.columns = df.iloc[0]
```

```python
df = df.drop(index=['Country/Region'])
```

## Visualize Examples

### This is the total number of aggregate global cases:

```python
df.sum(axis=1).plot()
```

```python
df0 = df[['Germany','Italy']]
```

### Periodicity
Since Covid seems to spread globally from the existing clusters, similar countries should have similar numbers, offset by the travel time of the virus epicenter. Observe the similarity between developments in Germany and Italy, when Italy is being offset by 8 days:

```python
# Difference Italy-Germany should be around 8 days

ax = df0['Italy'].shift(periods=8).plot(legend=True)
df0['Germany'].plot(ax=ax,legend=True)

ax.legend(['ITALY','GERMANY'])
```

## Try fitting a mathematical curve
The partial samples of the timeseries already available allow the fitting of a curve, presumably a model which grows exponentially and finally saturates on a maximum threshold. This assumption seems plausible given the real spread of a disease.

## First try: Exponential model

```python
from scipy.optimize import curve_fit

# This is the proposed function which we'll fit
def exp_fn(x,a,k,b): return a*np.exp(k*x+b)
```

### Germany:

```python
# X-axis ticks along with available time series, Y-axis captures measured data

x = np.arange(0,df0['Germany'].values.shape[0])
y = df0['Germany'].values
```

```python
# Optimiziation/fitting step Germany
popt_exp,pcov_exp = curve_fit(exp_fn,x,y.astype(np.float32),p0=[1,0,1])
```

```python
# Forecast Germany with optimal perimeters
df0['Germany Model'] = exp_fn(x,popt_exp[0],popt_exp[1],popt_exp[2])
```

### Italy

```python
# X-axis ticks along with available time series, Y-axis captures measured data

x = np.arange(0,df0['Italy'].values.shape[0])
y = df0['Italy'].values
```

```python
# Optimiziation/fitting step Germany
popt_exp,pcov_exp = curve_fit(exp_fn,x,y.astype(np.float32),p0=[1,0,1])
```

```python
# Forecast Germany with optimal perimeters
df0['Italy Model'] = exp_fn(x,popt_exp[0],popt_exp[1],popt_exp[2])
```

### Results
Observe the close fit for both Germany and Italy especially in the initial stages of the pandemic:

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(15,5))

df0[['Germany','Germany Model']].plot(ax=ax[0])
df0[['Italy','Italy Model']].plot(ax=ax[1])

plt.show()
```

## Second model: Sigmoid

```python
# Since exponential model cannot cover saturation, give sigmoid a try
def sigmoid_fn(x,k,b,T): return T*(np.exp(k*x+b))/(1+np.exp(k*x+b))
```

### Germany

```python
# Run curve-fitting process

x = np.arange(0,df0['Germany'].values.shape[0])
y = df0['Germany'].values

popt_sigmoid,pcov_sigmoid = curve_fit(sigmoid_fn,x,y.astype(np.float32),p0=[0,0,1000])
df0['Germany Sigmoid Model'] = sigmoid_fn(x,popt_sigmoid[0],popt_sigmoid[1],popt_sigmoid[2])
```

#### The optimal curve for Germany
The optimal curve with Germany parameters lets me conjecture that a total of around 4M poeple will be afflicted

```python
# Plot the optimal curve
plt.plot(np.arange(0,1000),sigmoid_fn(np.arange(0,1000),popt_sigmoid[0],popt_sigmoid[1],popt_sigmoid[2]))
```

### Italy

```python
# Run curve-fitting process

x = np.arange(0,df0['Italy'].values.shape[0])
y = df0['Italy'].values

popt_sigmoid,pcov_sigmoid = curve_fit(sigmoid_fn,x,y.astype(np.float32),p0=[0,0,1000])
df0['Italy Sigmoid Model'] = sigmoid_fn(x,popt_sigmoid[0],popt_sigmoid[1],popt_sigmoid[2])
```

#### Italy optimal curve
For Italy the total case count can be expected to be slightly above the Germany count, around 4.5M cases in total.

```python
# Plot the optimal curve
plt.plot(np.arange(0,1000),sigmoid_fn(np.arange(0,1000),popt_sigmoid[0],popt_sigmoid[1],popt_sigmoid[2]))
```

### Visualization of Results
Observe the still very close fit for the initial stages in both Germany and Italy. Tracing the fitted curve further down the line can yield an estimate for the total number of expected Covid cases in the first wave!

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(15,5))

df0[['Germany','Germany Sigmoid Model']].plot(ax=ax[0])
df0[['Italy','Italy Sigmoid Model']].plot(ax=ax[1])

plt.show()
```
