```python
import numpy as np
from numpy import random as rnd
from matplotlib import pyplot as plt
import datetime,os,warnings,csv

import tensorflow as tf
import pandas as pd
import seaborn as sns

from keras.models import Model,Sequential
from keras.layers import Input,Dense,LSTM,advanced_activations,Dropout,BatchNormalization,concatenate,Embedding

from sklearn.model_selection import train_test_split
from sklearn import preprocessing as pp
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor

sns.set()
warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('..\\..\\TF_data\\KEPLER\\exoTrain.csv',infer_datetime_format=True)
df_test = pd.read_csv('..\\..\\TF_data\\KEPLER\\exoTest.csv',infer_datetime_format=True)
```

```python
# Label 1: No exo-planet
# Label 2: Exo-planet

df_train.drop(['LABEL'],axis=1).loc[0].plot()
```

```python
ID = 10

labelID = df_train['LABEL'].values[ID]
fluxID = df_train.drop(['LABEL'],axis=1).loc[ID].values
```

```python
plt.plot(np.fft.fft(fluxID))
if labelID==1: 
    plt.title('No planet') 
else: 
    plt.title('Planet')
```

```python
DATA_train = []
LABELS_train = []

for k in range(df_train.shape[0]):
    print('Processing training sample %d/%d'%(k,df_train.shape[0]))
    
    l = df_train['LABEL'].values[k]
    d = df_train.drop(['LABEL'],axis=1).loc[k].values
    D = np.fft.fft(fluxID)
    
    DATA_train.append(D)
    LABELS_train.append(l)
    
DATA_train = np.stack(DATA_train,axis=0)
LABELS_train = np.stack(LABELS_train,axis=0)

DATA_test = []
LABELS_test = []

for k in range(df_test.shape[0]):
    print('Processing training sample %d/%d'%(k,df_test.shape[0]))
    
    l = df_train['LABEL'].values[k]
    d = df_train.drop(['LABEL'],axis=1).loc[k].values
    D = np.fft.fft(fluxID)
    
    DATA_test.append(D)
    LABELS_test.append(l)
    
DATA_test = np.stack(DATA_test,axis=0)
LABELS_test = np.stack(LABELS_test,axis=0)
```

```python
import pickle

pickle.dump(DATA_train,open('..\\..\\TF_data\\KEPLER\\DATA_train.p','wb'))
pickle.dump(LABELS_train,open('..\\..\\TF_data\\KEPLER\\LABELS_train.p','wb'))

pickle.dump(DATA_test,open('..\\..\\TF_data\\KEPLER\\DATA_test.p','wb'))
pickle.dump(LABELS_test,open('..\\..\\TF_data\\KEPLER\\LABELS_test.p','wb'))
```
