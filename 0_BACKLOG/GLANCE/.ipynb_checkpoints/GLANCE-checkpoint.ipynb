```python
import numpy as np
from numpy import random as rnd
from matplotlib import pyplot as plt
import tensorflow as tf
import os,warnings,datetime,csv

import pandas as pd

from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout
from keras.datasets import mnist

from sklearn.preprocessing import OneHotEncoder
```

```python
(xtrain,ytrain),(xtest,ytest) = mnist.load_data()

xtrain = xtrain/255
xtest = xtest/255

xtrain = np.reshape(xtrain, [xtrain.shape[0],xtrain.shape[1],xtrain.shape[2],1])
xtest = np.reshape(xtest, [xtest.shape[0],xtest.shape[1],xtest.shape[2],1])

ohe = OneHotEncoder(sparse=False)
ytrain = ohe.fit_transform(np.reshape(ytrain, [-1,1]))
ytest = ohe.transform(np.reshape(ytest, [-1,1]))
```

```python
mid = 215

plt.imshow(xtrain[mid,:,:,0])
plt.title(ytrain[mid,:])
plt.show()
```

```python
tf.reset_default_graph()

x = tf.placeholder(dtype=tf.float32, shape=[None,28,28,1], name='nn_input')
y = tf.placeholder(dtype=tf.int32, shape=[None,10], name='nn_output')

t = tf.layers.conv2d(x,filters=32,kernel_size=[5,5],activation=tf.nn.relu)
t = tf.layers.max_pooling2d(t,pool_size=[2,2],strides=[2,2])
t = tf.layers.conv2d(t,filters=64,kernel_size=[5,5],activation=tf.nn.relu)
t = tf.layers.max_pooling2d(t,pool_size=[2,2],strides=[2,2])
t = tf.reshape(t,[-1,int(t.shape[1])*int(t.shape[2])*int(t.shape[3])])
t = tf.layers.dense(t,units=1024,activation=tf.nn.relu)
t = tf.layers.dropout(t,rate=0.4)

logits = tf.layers.dense(t,units=10)
ppred = tf.nn.softmax(logits=logits)
cls = tf.argmax(ppred,axis=1)

loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits,labels=y))

opt = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)
```

```python
num_epochs = 1
batch_len = 10
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for e in range(num_epochs):
        print('Epoch %d...'%e)
        
        closs = []
        for k in range(xtrain.shape[0]//batch_len):
            
            xc = np.reshape(xtrain[k*batch_len:(k+1)*batch_len,:,:,:],[batch_len,28,28,1])
            yc = np.reshape(ytrain[k*batch_len:(k+1)*batch_len],[batch_len,10])
            
            _,cclass,lc = sess.run([opt,cls,loss],feed_dict={
                x:xc,
                y:yc
            })
            closs.append(lc)
            if k%100==0:
                print('Pred class: %s\nReal class: \n%s --> Loss=%s'%(str(cclass),str(yc),str(lc)))
        
    
    saver = tf.train.Saver()
    savestr = 'data\\mnist_cnn_'+datetime.datetime.today().strftime('%H-%M--%d-%m-%Y')+'_model.ckpt'
    save_path = saver.save(sess, savestr)
```
