```python
tf.reset_default_graph()

nn_in = tf.placeholder(dtype=tf.float32,shape=(None,25,12))
nn_out_exp = tf.placeholder(dtype=tf.float32,shape=(None,37))

x = tf.contrib.layers.conv2d(
    inputs=nn_in,
    num_outputs=200,
    kernel_size=3,
    stride=2,
    activation_fn=tf.nn.relu
)
x = tf.contrib.layers.flatten(
    inputs=x
)
nn_out = tf.contrib.layers.fully_connected(
    inputs=x,
    num_outputs=37,
    activation_fn=tf.nn.relu
)
```

```python
class GAME():
    
    def __init__(self,pixel_size=4,disp_size=(100,50),clocktick=1000,deathsum=100):
        self.pixel_size = pixel_size
        self.disp_size = disp_size
        self.clocktick = clocktick
        self.deathsum = deathsum
        
    def play_game(self,player,sess):
        screen =  pygame.display.set_mode(self.disp_size,RESIZABLE)
        clock = pygame.time.Clock()
        MAP = np.zeros(shape=[int(screen.get_width()//self.pixel_size),int(screen.get_height()//self.pixel_size)])
        MAP[0:10,0:10] = rnd.randint(0,2,(10,10))

        iterations = 0
        mapsum = []
        
        X = []
        Y = []
        try:
            running = True
            while running==True:
                screen.fill(0)
                for k in zip(np.where(MAP!=0)[0],np.where(MAP!=0)[1]):
                    pygame.draw.rect(screen,(255,255,0),(self.pixel_size*k[0],self.pixel_size*k[1],5,5),0)
                pygame.display.flip()

                MAP = self.update_map(MAP)
                mapsum.append(MAP.sum())
                MAP,nn_input,nn_output = player.change_map(MAP,sess)
                X.append(nn_input)
                Y.append(nn_output)
                
                if mapsum[-1]>self.deathsum or mapsum[-1]==0:running=False
                if any([e.type==pygame.KEYDOWN for e in pygame.event.get()]):running=False
                clock.tick(self.clocktick)
                iterations +=1
            pygame.quit()
        except SystemExit:
            pygame.quit()
        return iterations,mapsum,X,Y
    
    def update_map(self,MAP):
        for k in range(MAP.shape[0]):
            for kk in range(MAP.shape[1]):
                try:
                    neighbor_sum = MAP[k+1,kk]+MAP[k-1,kk]+MAP[k,kk+1]+MAP[k,kk-1]+MAP[k+1,kk+1]+MAP[k-1,kk-1]+MAP[k-1,kk+1]+MAP[k+1,kk-1]
                    if MAP[k,kk]==1:
                        if neighbor_sum<2 or neighbor_sum>3:
                            MAP[k,kk]=0
                    if MAP[k,kk]==0:
                        if neighbor_sum==3:
                            MAP[k,kk]=1
                except:
                    continue  
        return MAP
```

```python
class PLAYER():
    
    def __init__(self,learning_rate=0.01,discount=0.5):
        self.learning_rate = learning_rate
        self.discount = discount
    
    def change_map(self,MAP,sess):
        nn_input = MAP.copy().reshape(-1,25,12)
        nn_output = sess.run([nn_out],feed_dict={
            nn_in:nn_input
        })
        return MAP,nn_input,nn_output
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    player = PLAYER(sess)
    game = GAME()
    survived_iter,mapsum,X,Y = game.play_game(player,sess)
```
