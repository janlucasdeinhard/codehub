```python
import numpy as np
from numpy import random as rnd
from matplotlib import pyplot as plt
import os,warnings,datetime,time,sys,itertools

import torch
from torch import functional as F

import pandas as pd
```

```python
# Root path

basepath = '../../../../../../../../TF_data/ADMIN/v2/'
```

```python
# Load single transactions and concatenate

path = basepath + '01_SingleTransactions/'

df0 = pd.DataFrame()

# Read CSVs
for file in [f for f in os.listdir(path) if f.lower().endswith('.csv')]:
    tt = pd.read_csv(path+file,sep=';',encoding='latin-1')
    tt['Origin'] = file
    tt['Betrag'] = tt['Betrag'].apply(lambda x: str(x)[:-2]+'.'+str(x)[-2:] if ',' not in str(x) and '.' not in str(x) else str(x).replace(',','.'))
    df0 = pd.concat([df0,tt])
    
# Read XLSs
for file in [f for f in os.listdir(path) if f.lower().endswith('.xlsx')]:
    tt = pd.read_excel(path+file)
    tt['Origin'] = file
    tt['Betrag'] = tt['Betrag'].apply(lambda x: str(x)[:-2]+'.'+str(x)[-2:] if ',' not in str(x) and '.' not in str(x) else str(x).replace(',','.'))
    df0 = pd.concat([df0,tt])
```

```python
# Convert amt column to float
df0['Betrag'] = df0['Betrag'].astype(np.float)
```

```python
# Special case handeling: Create dictionary
usage_dict = dict()
for k in df0['Verwendungszweck'].unique():
    if 'kieser' in k.lower():
        usage_dict[k] = 'Kieser Beitrag'
    elif k.startswith('EREF+'):
        usage_dict[k] = k.replace('EREF+','')
    else:
        usage_dict[k] = k
        
# Apply special case dictionary
df0['Verwendungszweck'] = df0['Verwendungszweck'].apply(lambda x: usage_dict[x] if x in usage_dict.keys() else x)
```

```python
# Remove duplicate entries
df0 = df0.reset_index(drop=True)
df0 = df0.loc[df0.drop(columns=['Origin']).drop_duplicates().index.tolist()]
```

```python
must_match_cols = [
     'Auftragskonto'
    ,'Buchungstag'
    ,'Valutadatum'
    ,'Buchungstext'
    ,'Verwendungszweck'
    ,'Betrag'
    ,'Waehrung'
]
```

```python
df0 = df0.reset_index(drop=True)

t0 = df0[must_match_cols]
t0['Counter'] = 1

t0 = t0.groupby(must_match_cols).sum().reset_index().sort_values(by='Counter',ascending=False)
potential_duplicates = t0[t0['Counter']>1].drop(columns=['Counter']).apply(lambda x: '|'.join([str(k) for k in x]),axis=1).tolist()
```

```python
df0['Dupl_Key'] = df0[must_match_cols].apply(lambda x: '|'.join([str(k) for k in x]),axis=1)

df00 = df0[df0['Dupl_Key'].isin(potential_duplicates)].drop(columns=['Dupl_Key'])
df01 = df0[~df0['Dupl_Key'].isin(potential_duplicates)].drop(columns=['Dupl_Key'])
```

```python
df00['Potential Duplicate (Flag)'] = 'Y' 
df01['Potential Duplicate (Flag)'] = 'N' 
```

```python
df0 = pd.concat([df00,df01])
```

```python
# Load ground truth

path = basepath + '02_AccountTotal/'

df1 = pd.read_excel(path+'AccountTracker.xlsx')
```

```python
a0 = df0.copy()
a1 = df1.copy()

a0['Date yyyy-mm'] = a0['Buchungstag'].dt.strftime('%Y-%m')
a1['Date yyyy-mm'] = a1['Date'].dt.strftime('%Y-%m')
```

```python
a0 = a0[['Date yyyy-mm','Betrag']].groupby(['Date yyyy-mm']).sum()
```

```python
a1 = a1[['Date yyyy-mm','Account Total']].groupby(['Date yyyy-mm']).sum()
```

```python
a0['PrevMonth'] = (pd.to_datetime(a0.index) + pd.DateOffset(months=-1))
a0['PrevMonth'] = a0['PrevMonth'].dt.strftime('%Y-%m')
```

```python
pf = a1.reset_index().merge(a0,left_on='Date yyyy-mm',right_on='PrevMonth',how='left').drop(columns=['PrevMonth']).set_index('Date yyyy-mm')
```

```python
pf['Pred Total'] = pf[['Account Total','Betrag']].apply(lambda x: x[0]+x[1],axis=1).shift(1)
```

```python
pf
```

```python
pf['error'] = pf[['Account Total','Pred Total']].apply(lambda x: x[0]-x[1],axis=1)
```

```python
pf['error'].plot(rot=75)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(pf['error'].fillna(0),order=(3,1,0))
model_fit = model.fit()
print(model_fit.summary())
```

```python
plt.plot(model_fit.forecasts_error[0])
```

```python
pf['ARIMA Forecast'] = model_fit.forecasts[0]
pf['ARIMA Forecast Error'] = model_fit.forecasts_error[0]
```
