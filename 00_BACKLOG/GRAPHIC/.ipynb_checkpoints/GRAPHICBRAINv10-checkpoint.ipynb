```python
def div(x,y):
    res = np.abs(x-y)**2
    return res

def relu(x):
    res = 0
    if x>=0: res = x
    return res

def compute_energy(C,h):
    En_list = []
    for n in range(C.shape[0]):
        En = 0
        for m in range(C.shape[1]):
            En += C[n,m]*div(h[n],h[m])
        En /= np.count_nonzero(C[n,:])  
        En_list.append(En)
    E = 0
    for En in En_list:
        E += relu(En)
    E /= C.shape[0]
    return E

def compute_energy_gradient(C,h):
    gradient = []
    for n0 in range(C.shape[0]):
        Nn0 = np.where(C[n0,:]!=0)[0]
        num_neigh_n0 = Nn0.shape[0]
        dE = 0
        for nn in Nn0:
            num_neigh_nn = np.where(C[nn,:]!=0)[0].shape[0]
            Cnn = C[n0,nn]
            dE += Cnn*((h[n0]-h[nn])/num_neigh_n0 - (h[nn]-h[n0])/num_neigh_nn)
        dE *= 2/C.shape[0]
        gradient.append(dE)
    gradient = np.array(gradient)
    return gradient

def settle_network(C,h,hold_nodes=[],learning_rate = 0.85,thresholdE = 1e-4):
    Eprev = np.inf
    deltaE = np.inf
    while deltaE>thresholdE:
        E = compute_energy(C,h)
        deltaE = np.abs(E - Eprev)
        Eprev = E
        dh = compute_energy_gradient(C,h)
        for kk in hold_nodes:
            dh[kk] = 0
        h -= learning_rate*dh
    return C,h 

def make_random_network(nodes,perc_zeros=0.95):
    C = rnd.rand(nodes,nodes)
    h = rnd.randn(nodes)
    for i in range(nodes):
        for j in range(nodes):
            if i==j:
                C[i,j]=0
    for i in range(nodes):
        for j in range(nodes):
            if rnd.rand()<perc_zeros:
                C[i,j]=0                  
    for i in range(nodes):
        for j in range(i,nodes):
            C[j,i] = C[i,j]
    for i in range(C.shape[0]):
        if np.count_nonzero(C[i,i:])==0:
            C[i,rnd.choice(range(i,C.shape[1]))]=1  
            C[rnd.choice(range(i,C.shape[1])),i]=1        
    return C,h

def train(nn_in,nn_out,C,h,in_nodes,out_nodes):
    if not (len(nn_in)==len(in_nodes))&(len(nn_out)==len(out_nodes)): raise Exception('Wrong input or output shape')
    C,h = settle_network(C,h,hold_nodes=[])
    for k in range(len(nn_in)):
        h[in_nodes[k]] = nn_in[k]
    for k in range(len(nn_out)):
        h[out_nodes[k]] = nn_out[k]
    C,h = settle_network(C,h,hold_nodes=in_nodes+out_nodes)
    
    # TODO change graph
    
    C,h = settle_network(C,h,hold_nodes=[])
    return C,h

def predict(nn_in,C,h,in_nodes,out_nodes):
    if not len(nn_in)==len(in_nodes): raise Exception('Wrong input shape')
    C,h = settle_network(C,h,hold_nodes=[])
    for k in range(len(nn_in)):
        h[in_nodes[k]] = nn_in[k]
    C,h = settle_network(C,h,hold_nodes=in_nodes)
    nn_out = []
    for node in out_nodes:
        nn_out.append(h[node])
    C,h = settle_network(C,h,hold_nodes=[])
    return nn_out
```

```python
C,h = make_random_network(5,perc_zeros=0.6)
G = nx.from_numpy_array(C)

nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```
