```python
import numpy as np 
from numpy import random as rnd
from matplotlib import pyplot as plt
import sys,os,datetime,warnings

import tensorflow as tf
import pandas as pd
import seaborn as sns

from keras.models import Model,Sequential
from keras.layers import Dense,LSTM,Input,BatchNormalization,Conv2D,Conv2DTranspose,Activation,Concatenate,Bidirectional,CuDNNGRU,Dropout,Dot,Flatten,Embedding

from sklearn.model_selection import train_test_split

from scipy.stats import norm

sns.set()
warnings.filterwarnings('ignore')
```

```python
# Creates a randomly shaped PDF from aggregated Gaussians
# Provides methods to sample the PDF
class RANDOM_PDF_GENERATOR():
    # Initialize
    def __init__(self,N_MAX=2,MU_MIN=-5,MU_MAX=5,SIGMA_MIN=0.1,SIGMA_MAX=5):
        self.N_MAX = N_MAX
        self.MU_MIN = MU_MIN
        self.MU_MAX = MU_MAX
        self.SIGMA_MIN = SIGMA_MIN
        self.SIGMA_MAX = SIGMA_MAX
        self.initialized = False
        self.generate_parameters()
        return
    # Once create random parameters
    def generate_parameters(self):
        self.n_gaussians = rnd.randint(1,self.N_MAX) 
        self.mu_set = rnd.rand(self.n_gaussians)*(self.MU_MAX - self.MU_MIN) + self.MU_MIN
        self.sigma_set = rnd.rand(self.n_gaussians)*(self.SIGMA_MAX - self.SIGMA_MIN) + self.SIGMA_MIN
        self.initialized = True
        return
    # Evaluate CDF at point x
    def eval_cdf(self,x):
        y = 0
        for k in range(self.n_gaussians):
            y += (self.sigma_set[k]*norm.cdf(x+self.mu_set[k]))
        return y 
    # Sample PDF with multiple samples
    def sample_pdf(self,num_samples=1000):
        x = np.arange(self.MU_MIN-5,self.MU_MAX+5,0.01)
        y = self.eval_cdf(x)
        s = [x[np.argmin((y - ((max(y) - min(y))*rnd.rand()+min(y)))**2)] for k in range(num_samples)]
        return s
```

```python
# Creates conditional distribution Pr[Y|X] based on binning X; X is passed as kernel distribution g
# Provides methods to sample the conditional PDF
class RANDOM_CONDITIONAL_PDF():
    # Initialize
    def __init__(self,g,NUMBINS=10,PERC_VARIATION=0.3):
        self.g = g
        self.NUMBINS = NUMBINS
        self.PERC_VARIATION = PERC_VARIATION
        self.i_width_real = (self.g.MU_MAX+self.g.SIGMA_MAX) - (self.g.MU_MIN-self.g.SIGMA_MIN)
        self.i_max = (self.g.MU_MAX+self.g.SIGMA_MAX) + self.i_width_real*self.PERC_VARIATION
        self.i_min = (self.g.MU_MIN-self.g.SIGMA_MIN) - self.i_width_real*self.PERC_VARIATION
        self.THRESHOLDS = rnd.rand(self.NUMBINS)*(self.i_max-self.i_min) + self.i_min
        self.THRESHOLDS.sort()
        self.p_cond = [RANDOM_PDF_GENERATOR() for k in range(self.NUMBINS+1)]
        return
    # Sample conditional distribution with original distribution input
    def sample_cond(self,X_o):
        B = []
        for x in X_o:
            b = 0
            for k in range(len(self.THRESHOLDS)):
                if x<self.THRESHOLDS[k]: break
                b += 1
            B.append(b)
        Y = [self.p_cond[b].sample_pdf(1)[0] for b in B]
        return Y
    # Sample PDF with multiple samples
    def sample_pdf(self,num_samples=1000):
        X_o = self.g.sample_pdf(num_samples)
        B = []
        for x in X_o:
            b = 0
            for k in range(len(self.THRESHOLDS)):
                if x<self.THRESHOLDS[k]: break
                b += 1
            B.append(b)
        Y = [self.p_cond[b].sample_pdf(1)[0] for b in B]
        return Y
```

```python
# Hidden variable, generates correlated Y0,Y1
X = RANDOM_PDF_GENERATOR()
# Correlated observables Y0,Y1
Y0 = RANDOM_CONDITIONAL_PDF(X)
Y1 = RANDOM_CONDITIONAL_PDF(X)

# Causally related observables Z0,Z1
Z0 = RANDOM_PDF_GENERATOR()
Z1 = RANDOM_CONDITIONAL_PDF(Z0)
```

```python
sample_num = 10000

df = pd.DataFrame()
df['Y0'] = Y0.sample_pdf(sample_num)
df['Y1'] = Y1.sample_pdf(sample_num)
df['Z0'] = Z0.sample_pdf(sample_num)
df['Z1'] = Z1.sample_pdf(sample_num)
```

```python
sns.lmplot(x='Y0',y='Y1',data=df,fit_reg=False)
```

```python
sns.lmplot(x='Z0',y='Z1',data=df,fit_reg=False)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
x = df['Y0'].values.reshape(-1,1)
y = df['Y1'].values.reshape(-1,1)

x0,x1,y0,y1 = train_test_split(x,y,test_size=0.2)
```

```python
clf = MLPRegressor(solver='lbfgs',alpha=1e-5,hidden_layer_sizes=(150,5),random_state=1)
clf.fit(x0,y0)
```

```python
clf.score(x1,y1)
```
